```python
import pandas as pd
import matplotlib.pyplot as mp
import seaborn as sns

data = pd.read_csv('/content/Titanic-Dataset.csv')

print(data.info())

data.isnull().sum()

data['Sex']=data['Sex'].replace({'male':1,'female':0})

data['Age'].fillna(data['Age'].mean(), inplace=True)

Embarked_counts = data['Embarked'].value_counts()
print(Embarked_counts)

data["Embarked"].fillna(data["Embarked"].mode()[0], inplace=True)

data['Embarked'] = data['Embarked'].replace({'S':1,'C':2,'Q':3})

data=data.drop(columns=['PassengerId', 'Name', 'Cabin', 'Ticket'], axis=1)

Survived_counts = data['Survived'].value_counts()
print(Survived_counts)

age=sns.FacetGrid(data,col='Survived')
age.map(mp.hist,'Age',bins=20)
mp.show()
```

```python
sns.catplot(x ="Sex", hue ="Survived",
kind ="count", data = data)

mp.show()
```

```python
sibsp=sns.FacetGrid(data,col='Survived')
sibsp.map(mp.hist,'SibSp',bins=9)
mp.show()
```

```python
pclass=sns.FacetGrid(data,col='Survived')

pclass.map(mp.hist,'Pclass')
mp.show()
```

```python
pclass=sns.FacetGrid(data,col='Survived')

pclass.map(mp.hist,'Parch')
mp.show()
```

```python
mp.figure(figsize=(16, 10))
sns.heatmap(data.corr(), annot=True)
mp.show()
```

```python
sns.violinplot(x ="Sex", y ="Age", hue ="Survived",
data = data, split = True)
mp.show()
```

```python
sns.catplot(x ='Embarked', hue ='Survived',
kind ='count', col ='Pclass',data=data)

mp.show()
```
